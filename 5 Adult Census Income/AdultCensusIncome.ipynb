```python
columns = [
    "age", "workclass", "fnlwgt", "education", "education_num",
    "marital_status", "occupation", "relationship", "race", "sex",
    "capital_gain", "capital_loss", "hours_per_week", "native_country",
    "income"
]
```

```python
import pandas as pd

# Cargar datos de entrenamiento
train = pd.read_csv("adult.data", header=None, names=columns, sep=",", skipinitialspace=True)

# Cargar datos de prueba
test = pd.read_csv("adult.test", header=None, names=columns, sep=",", skipinitialspace=True, comment="|")

print(train.head())
print(test.head())
```

```python
train = train.replace("?", pd.NA)
test = test.replace("?", pd.NA)
```

```python
test["income"] = test["income"].str.replace(".", "", regex=False)
```
