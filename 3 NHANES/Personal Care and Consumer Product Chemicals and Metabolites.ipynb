```python
import pandas as pd

# Cargar archivo XPT
df = pd.read_sas("dataset.xpt", format="xport")
```

```python
# Ver nombres de columnas
print(df.columns)
```

```python
df.info()
```

```python
# columnas de concentraciones
cols_values = [col for col in df.columns if col.startswith("URX")]

# columnas de flags (LOD)
cols_flags = [col for col in df.columns if col.endswith("LC")]
```

```python
df[cols_flags].value_counts()
```

```python
import numpy as np

# convertir valores bajo LOD a NaN (o a LOD/√2 dependiendo del análisis)
for col in cols_values:
    flag_col = col.replace("URX", "URD") + "LC"
    if flag_col in df.columns:
        df.loc[df[flag_col] == 1, col] = np.nan
```

```python
df[cols_values].describe()
```
